```python
import pandas as pd

df = pd.read_excel("/home/cristian/projects/rag_pae/data/Info_docs_y_campos.xlsx", sheet_name=1)
df
```

```python
import pprint

total = ""
for fila in df['Campos requeridos']:
    total = total + fila + ", "
total_lst = total.split(",")

#pprint.pp((total_lst))

for i,campo in enumerate(total_lst):
    if campo[0] == " ":
        campo = campo[1:]
    if len(campo) == 0:
        total_lst.pop(i)
    
print(len(total_lst))


new_df = pd.DataFrame(total_lst, columns=["Campos requeridos"])
conteo = new_df["Campos requeridos"].value_counts().reset_index()
conteo.columns = ["Campo", "Conteo"]
conteo = conteo.sort_values(by="Conteo", ascending=False)

selected_fields = conteo[conteo["Conteo"] > 5]["Campo"]
selected_fields.reset_index(drop=True)

print(conteo)
```
